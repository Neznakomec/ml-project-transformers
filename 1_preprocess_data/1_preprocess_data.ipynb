```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pickle import load,dump
from numpy import array
from collections import defaultdict
import os
```

```python
path = './text/'
df = pd.read_table(path + 'Flickr8k.lemma.token.txt', header=None)
df.columns = ['name', 'description']
df[['pic_name', 'ext_num', 'trash']] = df['name'].str.split('.',expand=True)
```

```python
df
```

```python
df1 = df[['pic_name', 'description']]
df1
```

```python
dictionary = defaultdict(list)
for row in df1.iterrows():
    key_pic_name = row[1][0]
    value_pic_desc = row[1][1].lower()
    dictionary[key_pic_name].append(value_pic_desc)
```

```python
import json
data = json.loads(json.dumps(dictionary))
data
```

```python
# check some picture
data['2394267183_735d2dc868']
```

```python
# split descriptions into tokens
for img_i in data:
    for caption_i in range(len(data[img_i])):
        sentence = data[img_i][caption_i]
        data[img_i][caption_i] = ["#START#"] + sentence.split(' ') + ["#END#"]
```

```python
data
```

```python
with open('./output/captions_tokenized.json', 'w') as f:
    json.dump(data, f)
```
